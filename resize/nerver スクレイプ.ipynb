```python
url = "https://matome.naver.jp/odai/2139355920484426101"
import requests
from bs4 import BeautifulSoup
from pprint import pprint
import re
import pandas
```

```python
idx = 4
```

```python
outputdata = []
for i in range(1,125):
    resp = requests.get(url + "?page=" + str(i))
    soup = BeautifulSoup(resp.text,  "lxml")
    datas = soup.find_all("div", {"class":"mdMTMWidget01Content01"})
    for data in datas:
        _data = []
        img = data.img
        if img is None:
            continue
        src = img["src"]
        pprint(src)
        response = requests.get(src, allow_redirects=False, timeout=10)
        image = response.content
        with open("./input/"+str(idx)+".jpg", "wb") as fout:
                fout.write(image)
                idx += 1


        tmp = data.text.replace("\t", "\n").replace("\r", "\n").replace(" ", "\n")
        textarr = re.split("\n+", tmp.strip())
        _data.append({"tags":textarr, "id":idx-1, "extension":"jpg"})    
        outputdata.extend(_data)
    
```

```python
pd = pandas.DataFrame(outputdata)
```

```python
with open('df.json', 'w', encoding='utf-8') as file:
    pd.to_json(file, force_ascii=False, orient='records')
```
